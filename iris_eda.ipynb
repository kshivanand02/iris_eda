```python
#Importing required lobraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For inline plotting
%matplotlib inline
```

```python
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)

# Display the first few rows
df.head(15)
```

```python
# Mean of numeric columns
mean_df = df.mean(numeric_only=True).to_frame(name='Mean')
print("Mean:")
display(mean_df)

# Median of numeric columns
median_df = df.median(numeric_only=True).to_frame(name='Median')
print("Median:")
display(median_df)

# Standard Deviation
std_df = df.std(numeric_only=True).to_frame(name='Standard Deviation')
print("Standard Deviation:")
display(std_df)

# Describe - Summary statistics
print(".describe() output:")
display(df.describe())
```

```python
# Set plot size
df.hist(figsize=(15, 10), bins=20, edgecolor='black', color='skyblue')
plt.suptitle("Histograms of All Numeric Features", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
